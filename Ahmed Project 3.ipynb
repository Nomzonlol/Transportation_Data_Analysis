```python
import pandas as pd
import matplotlib.pyplot as plt

# Load NHTS data
nhts_df = pd.read_csv('NHTS.csv')

# Load NGSIM data
ngsim_df = pd.read_csv('NGSIM.csv')
```

```python
import matplotlib.pyplot as plt

# Count the number of vehicles for each fuel type
fuel_counts = nhts_df['fuel_type'].value_counts()

# Create the bar chart
plt.figure(figsize=(10, 6))
fuel_counts.plot(kind='bar', color='red', edgecolor='black')

# Add professional labels and title
plt.title('Frequency of Vehicle Fuel Types (NHTS Data)')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Vehicles')

# Rotate x-axis labels so they are easy to read
plt.xticks(rotation=45)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Clear any previous figures
plt.figure(figsize=(10, 6))

# Only run the Seaborn version
sns.histplot(nhts_df['vehicle_age'], bins=15, kde=True, color='purple')

# Add labels and title
plt.title('Distribution of Vehicle Age')
plt.xlabel('Age of Vehicle (Years)')
plt.ylabel('Number of Vehicles (Count)')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Using a completely different variable: vehicle_year
nhts_df.boxplot(column='vehicle_year', by='urban', figsize=(10, 6))

# Add labels and title
plt.title('Distribution of Vehicle Model Year by Urban vs. Rural Area')
plt.suptitle('') 
plt.xlabel('Area Type (Urban/Rural)')
plt.ylabel('Vehicle Model Year')

plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Filter the data for JUST one vehicle (Trajectory #1)
# You can change '1' to any number that exists in your data
single_car = ngsim_df[ngsim_df['trajectory_number'] == 1].sort_values(by='Time')

# 2. Plot the Leader Speed for that one car
plt.figure(figsize=(10, 5))
plt.plot(single_car['Time'], single_car['leader_speed(m/s)'], color='blue', label='Lead Vehicle')
plt.title('Time-Series: Leader Speed (Vehicle Trajectory #1)')
plt.xlabel('Time (seconds)')
plt.ylabel('Speed (m/s)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# 3. Plot the Follower Speed for that one car
plt.figure(figsize=(10, 5))
plt.plot(single_car['Time'], single_car['follower_speed(m/s)'], color='green', label='Following Vehicle')
plt.title('Time-Series: Follower Speed (Vehicle Trajectory #1)')
plt.xlabel('Time (seconds)')
plt.ylabel('Speed (m/s)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
